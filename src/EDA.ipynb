## Data Cleaning

We want fill missing data instead of dropping them . One way to do this is by filling the mean values of the column .

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

#### Conveted data types where needed

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

```python
for col in df.columns:
    if df[col].dtype == 'object' and col != 'timestamp':
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

The diagonal  line is perfectly red with values 1.0 , which simply reflects that each feature is perfectly correleted with itself. Most off-diagonal values are light blue or near zero ,indicating low correlation between most features. Some zone's humidity and temperature pairs may have have slight relationship ,but generally : 
1. No sttrong multicollinearity is present
2. This is good for regression model because it reduces redundancy among predictors.
